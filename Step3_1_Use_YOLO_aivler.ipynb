```python
import os
import zipfile
```

```python
dataset1 = os.path.join('/content/drive/MyDrive/project4/Datasets/YOLO/Face_Recognition_1.zip' )
dataset2 = os.path.join('/content/drive/MyDrive/project4/Datasets/YOLO/Face_Recognition_2.zip' )

dataset1, dataset2
```

```python
## Colab에 생성할 본인 얼굴 폴더 경로
extract_folder = '/content/temp_dataset'

## 위의 경로에 폴더가 없을 때 생성
if not os.path.exists(extract_folder) :
    os.makedirs(extract_folder)

for dataset in [dataset1, dataset2] :
    ## 위의 경로에 압축을 해제
    with zipfile.ZipFile(dataset, 'r') as zip_ref :
        file_list = zip_ref.namelist()

        ## 세부 폴더 images 생성
        if not os.path.exists( os.path.join(extract_folder, 'images') ) :
            os.makedirs( os.path.join(extract_folder, 'images') )
        ## 세부 폴더 labels 생성
        if not os.path.exists( os.path.join(extract_folder, 'labels') ) :
            os.makedirs( os.path.join(extract_folder, 'labels') )

        ## file_list의 파일을 하나씩 불러온다
        for f in file_list :
            ## 파일명이 jpg로 끝난다면
            if not f.endswith('/') and f.lower().endswith('.jpg') :
                f_name = os.path.basename(f)
                d_path = os.path.join(extract_folder, 'images')
                d_path = os.path.join(d_path, f_name)

                with zip_ref.open(f) as source, open(d_path, 'wb') as target :
                    target.write(source.read() )

            ## 파일명이 txt로 끝난다면
            if not f.endswith('/') and f.lower().endswith('.txt') :
                f_name = os.path.basename(f)
                d_path = os.path.join(extract_folder, 'labels')
                d_path = os.path.join(d_path, f_name)

                with zip_ref.open(f) as source, open(d_path, 'wb') as target :
                    target.write(source.read() )
```

#### 2) 모델이 요하는 구조의 폴더 생성+ 이동

```python
## 원본 폴더 경로들
source_folders = {'images': '/content/temp_dataset/images',
                  'labels': '/content/temp_dataset/labels'
                  }

## 타겟 폴더 경로들
target_folders = {'images':{'train':'/content/Datasets/images/train',
                            'val':'/content/Datasets/images/val',
                            # 'test':'/content/Datasets/images/test',
                            },
                  'labels':{'train':'/content/Datasets/labels/train',
                            'val':'/content/Datasets/labels/val',
                            # 'test':'/content/Datasets/labels/test',
                            }
                  }

## 폴더 생성 함수
def create_folder(path):
    if not os.path.exists(path):
        os.makedirs(path)

## Datasets 폴더 생성
create_folder('/content/Datasets')

## 하위 폴더 생성 : images, labels 및 train/val
for _, folders in target_folders.items() :
    for value in folders.values() :
        create_folder(value)
```

#### 3) 각 폴더에 데이터 옮기기

```python
## 데이터 스플릿 비율
split_size = 0.8

## images, labels 파일을 정렬하여 리스트화
image_list = sorted(glob.glob(os.path.join(source_folders['images'], '*')))
label_list = sorted(glob.glob(os.path.join(source_folders['labels'], '*')))

## 묶고 섞고 다시 리스트화
combined = list(zip(image_list, label_list))
random.shuffle(combined)
image_list, label_list = zip(*combined)

## 총 파일 수 계산
total_files = len(image_list)
tr_val_split = int(total_files * split_size)

print('전체 파일 수 : ', total_files)
print('training set 파일 수 : ', tr_val_split)
print('validation set 파일 수 : ', total_files - tr_val_split)

## train / val로 분리
tr_images, tr_labels = image_list[ : tr_val_split], label_list[ : tr_val_split]
val_images, val_labels = image_list[tr_val_split : ], label_list[tr_val_split : ]

print( 'training set 이미지 수 : ', len(tr_images), ' training set 라벨 수 : ', len(tr_labels) )
print( 'validation set 이미지 수 : ', len(val_images), 'validation set 라벨 수 : ', len(val_labels) )
```

```python
## 파일 이동 함수
def move_files(file_list, target_folder):
    for file_path in file_list:
        shutil.copy(file_path, os.path.join(target_folder, os.path.basename(file_path)))

# train, val로 이미지 및 라벨 파일 옮기기
move_files(tr_images, target_folders['images']['train'])
print("train set 이미지 파일 이동 완료!")
move_files(tr_labels, target_folders['labels']['train'])
print("train set 라벨 파일 이동 완료!")

move_files(val_images, target_folders['images']['val'])
print("val set 이미지 파일 이동 완료!")
move_files(val_labels, target_folders['labels']['val'])
print("val set 라벨 파일 이동 완료!")
```
